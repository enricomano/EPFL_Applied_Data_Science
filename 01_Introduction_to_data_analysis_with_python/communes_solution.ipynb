# D. Probabilities

Compute the probability that a randomly selected commune with elevation over 2000 is from the canton of Valais.

Using the following formula:
$$
P(A|B)=\frac{P(A,B)}{P(A)}
$$

Where P(A,B) is the probability to have a commune from the canton of Valais with an elevation greater then 2000 m and P(A) is the probability to have a commune with an elevation greater then 2000 m

```python
P_A_and_B = df[(df.Canton == 'VS') & (df.Elevation >= 2000)].shape[0]/df.shape[0]
P_A = df[df.Elevation >= 2000].shape[0]/df.shape[0]
```

```python
P_B_given_A = P_A_and_B/P_A
print('The probability that a randomly selected commune with elevation over 2000 is from the canton of Valais is', 
      round(100*P_B_given_A, 2), 
      '%')
```
